### 对单个单元操作

```python
xw.Range('A1')
```

```python
import datetime as dt
sht = xw.sheets[0]
```

```python
sht.range('A1').value = 1
```

```python
sht.range('A2').value = 'Hello'
```

```python
sht.range('A3').value is None
```

```python
sht.range('A4').value = dt.datetime(2000, 1, 1)
```

### 列表 

```python
# 1d
sht.range('B1').value = [[1],[2],[3],[4],[5]]
```

```python
sht.range('C1').value = [1, 2, 3, 4, 5]
```

```python
sht.range('d1').options(transpose=True).value = [1,2,3,4]
```

```python
# 2d
sht.range('a1:A5').options(ndim=2).value
```

```python
sht.range('A10').value = [['Foo 1', 'Foo 2', 'Foo 3'], [10, 20, 30]]
```

### 扩展

```python
rng1.value
```

```python
rng2.value
```

### NumPy arrays

```python
import numpy as np
sht.range('A15').value = np.eye(3)
```

### Pandas DataFrames

```python
df = pd.DataFrame([[1.1, 2.2], [3.3, None]], columns=['one', 'two'])
```

```python
df
```

```python
sht.range('A19').value = df
```

```python
sht.range('A23').options(index=False).value = df
```

```python
sht.range('A27').options(index=False, header=False).value = df
```

### Pandas Series

```python
s = pd.Series([1.1, 3.3, 5., np.nan, 6., 8.], name='myseries')
```

```python
sht.range('A30').value = s
```
