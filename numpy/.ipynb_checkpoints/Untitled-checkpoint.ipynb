```python
import numpy as np
```

### 合并

```python
x = np.arange(100).reshape([10,10])
y = np.arange(10)
```

```python
training_data = np.hstack([x,y.reshape([-1,1])])
```

```python
training_data
```

### 随机排列

```python
np.random.shuffle(training_data)
training_data
```

### 切分

```python
x,y = np.array_split(training_data, [-1], axis=1)
x
```

### 随机序号数

```python
r = np.random.permutation(len(y))
r
```

```python
x, y = x[r,:], y[r]
x
```
