```python
from sklearn import svm
from sklearn import datasets
clf = svm.SVC()
iris = datasets.load_iris()
X, y = iris.data, iris.target
clf.fit(X, y)  
```

# 保存模型

## 使用 joblib

```python
from sklearn.externals import joblib
joblib.dump(clf, 'SVM_joblib.pkl') 
```

## 使用 pickle

```python
import pickle
with open("SVM_pickle.pkl", 'wb') as f:  
    pickle.dump(clf, f)
```

# 导入模型

## joblib 读入

```python
from sklearn.externals import joblib
clf_load = joblib.load('SVM_joblib.pkl') 
```

```python
clf_load.score(X, y)
```

## pickle 读入

```python
with open('SVM_pickle.pkl', 'rb') as f:
    clf_loar_pickle = pickle.load(f)
```

```python
clf_loar_pickle.score(X, y)
```
