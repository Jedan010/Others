```python
import seaborn as sns
from scipy.io import loadmat
from sklearn.metrics import f1_score
```

```python
data = loadmat('x8data1.mat')
list(data.keys())
```

```python
X = data['X']
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.covariance import EllipticEnvelope
cla = EllipticEnvelope()
cla.fit(X)
y_dis = cla.decision_function(X)
```

```python
plt.plot(y_dis)
```

```python
## threshold
threshold = np.percentile(y_dis, 1.7)
outliers = y_dis < threshold
```

```python
plt.scatter(X[outliers==False, 0], X[outliers==False, 1])
plt.scatter(X[outliers, 0], X[outliers, 1])

xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()
xx, yy = np.meshgrid(np.linspace(xmin, xmax, 500), np.linspace(ymin, ymax, 500))
Z = cla.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[threshold],linestyles='dotted', colors='r',)
```

```python
sns.distplot(X[outliers==False, 0], rug=True)
sns.rugplot(X[outliers, 0], color='r')
```
